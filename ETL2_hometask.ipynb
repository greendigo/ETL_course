Данные строки практически полностью дублируются, разница либо в названии клуба, либо в амлуа игрока

Удалим дубликаты

```python
df.drop_duplicates('ID', inplace=True)
```

Теперь необходимо унифицировать значения в названиях клубов, для этого нужно:

1) удалить все вхождения "1. " в названиях
2) устранить неоднозначность в сокращениях "Fc" и "FC", приведя к единому формату без сокращения (также учитывая возможные ошибки, связанные с русской раскладкой клавиатуры при написании буквы C)

```python
def club_edit(club):
    res = club.replace('1. ', '')
    res = res.replace('Fc ', '')
    res = res.replace('FС ', '')
    res = res.replace('FC ', '')
    return res
    
df['Club'] = df['Club'].apply(club_edit)
```

```python
df['Club'].sort_values().unique()
```

Теперь добавляем колонку для возрастной группы футболистов

```python
def age_category(age):
    if age<20:
        return 'under 20'
    elif 20<=age<30:
        return 'between 20 and 30'
    elif 30<=age<36:
        return 'between 20 and 36'
    else:
        return '36 and older'
    
df['Age_category'] = df['Age'].apply(age_category)
df[['Age', 'Age_category']]
```
